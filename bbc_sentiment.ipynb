```python
from langchain_community.llms import Replicate
import os
import pandas as pd
from tqdm import tqdm

os.environ["REPLICATE_API_TOKEN"] = "your_replicate_api_token_here"
model = "ibm-granite/granite-3.3-8b-instruct"
llm = Replicate(model=model, replicate_api_token=os.environ["REPLICATE_API_TOKEN"])
print("Model Granite siap digunakan.")
```

```python
from google.colab import files
import pandas as pd

uploaded = files.upload()
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)
print("Jumlah data:", len(df))
df.head()
```

```python
def classify_sentiment(text):
    prompt = f"""
    Determine if the following news text expresses a positive or negative sentiment.
    Respond only with one word: Positive or Negative.

    Text:
    {text}
    """
    try:
        return llm.invoke(prompt).strip()
    except Exception as e:
        return f"Error: {e}"
```

```python
import time
results = []
for text in tqdm(df["description"].head(10), desc="Classifying"):
    sentiment = classify_sentiment(text)
    results.append(sentiment)
    time.sleep(10)

df_results = df.head(10).copy()
df_results["predicted_sentiment"] = results
```

```python
pd.set_option("max_colwidth", 200)
display(df_results[["description", "predicted_sentiment"]])
```
